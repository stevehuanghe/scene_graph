```python
import pandas as pd

csv_file = './scores.csv'
data = pd.read_csv(csv_file, '\t')
data
```

```python
pids = list(data['predicate ID'])
names = list(data['Name'])
print(names)
freq = list(data['count'])
mtn_ce = list(data['MotifNet+CE'])
mtn_cce_rank = list(data['MotifNet+CCE+Rank'])
lkn_ce = list(data['LinkNet+CE'])
lkn_cce_rank = list(data['LinkNet+CCE+Rank'])
kern_ce = list(data['KERN'])
lkn_fc = list(data['LinkNet+Focal'])
kern_cce_rank = list(data['KERN+CCE+Rank'])

all_data = zip(freq, pids, names, mtn_ce, mtn_cce_rank, lkn_ce, lkn_cce_rank, kern_ce, lkn_fc, kern_cce_rank)
all_sorted = sorted(all_data, key=lambda x: -x[0])
freq, pids, names, mtn_ce, mtn_cce_rank, lkn_ce, lkn_cce_rank, kern_ce, lkn_fc, kern_cce_rank = zip(*all_sorted)
```

```python
def autolabel_all(ax, rects_list):
    """Attach a text label above each bar in *rects*, displaying its height."""
    cnt = 0
    idx = 0
    for rect in rects_list[0]:
        height = max([r[i].get_height() for r in rects_list])
        idx+=1
        if height == 0.0:
            height = (cnt%2)*3 + 3
#         height += 2
        ax.annotate('{}'.format(freq[cnt]),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        cnt += 1
```

```python
def get_stats(diff):
    cnt = np.sum(diff > 0)
    print('num increase: ', cnt)
    cnt = np.sum(diff[5:] > 0)
    print('num increase [5:]: ', cnt)
    cnt = np.sum(diff[10:] > 0)
    print('num increase [10:]: ', cnt)
    cnt = np.sum(diff[30:] > 0)
    print('num increase [30:]: ', cnt)
    

print('MotifNet')
get_stats(diffs1)
print('LinkNet')
get_stats(diffs2)
print('KERN')
get_stats(diffs3)
```
